This notebook is intended to update CTA's GeoJSON path data to facilitate color-coding train lines.
THis could be done using GeoPandas, but could also treat this like a regular JSON file

```python
import json
```

```python
# Read the GeoJSON file
with open("../cta-lines.geojson", "r") as file:
    data = json.load(file)

# assign color codes
color_mapping = {
    "Blue":"#00a1de",
    "Brown":"#62361b",
    "Green":"#009b3a",
    "Orange":"#f9461c",
    "Pink":"#e27ea6",
    "Purple":"#522398",
    "Red":"#c60c30",
    "Yellow":"#f9e300"
}

# Extract and display all names from the properties attribute
for feature in data['features']:
    name = feature['properties']['Name']
    color = name.split(",")[0].split()[0]
    feature['properties']['color'] = color
    if color in color_mapping:
        feature['properties']['color-hex'] = color_mapping[color]
    if 'description' in feature['properties']:
        del feature['properties']['description']
```

```python
for feature in data['features']:
    print (feature['properties']['color'])
```

```python
# (Optional) Write the updated data back to the original file
with open("../cta-lines-revised.geojson", "w") as file:
    json.dump(data, file, indent=4)  # 'indent=4' for formatted, readable output
```
